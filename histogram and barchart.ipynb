```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('births.csv')

plt.hist(data['births'], bins=5, color='blue', edgecolor='black')
plt.title('Distribution of Births')
plt.xlabel('births')
plt.ylabel('Gender')
plt.show()

gender_counts = data['gender'].value_counts()
gender_counts.plot(kind='bar', color='orange')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
